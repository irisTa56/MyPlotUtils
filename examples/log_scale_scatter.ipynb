```python
import tk_plot_utils as tk
```

```python
tk.init_plotly(connected=True)
```

```python
import numpy as np
```

## Exponential Function

```python
x_array = np.arange(0, 10, 0.01)
y_array = np.exp(x_array)
```

```python
scatter = tk.make_scatter({
    "x": x_array,
    "y": y_array,
    "mode": "markers",
    "marker": {"maxdisplayed": 50}
  })
plt = tk.plotly()
```

```python
plt.set_title("exponential")
plt.set_x_title("horizontal", "x")
plt.set_y_title("vertical", "exp(x)")
```

```python
plt.show(scatter)
```

Remove range and tick layouts for linear scale before changing logarithmic scale.

```python
plt.delete_axis_layout("y", "range")
plt.delete_axis_layout("y", "dtick")
```

```python
plt.set_axis_layout("y", "type", "log")
plt.set_axis_layout("y", "tickformat", "e")
```

```python
plt.show(scatter)
```

Use another exponential format.

```python
plt.delete_axis_layout("y", "tickformat")
```

```python
plt.set_axis_layout("y", "exponentformat", "power")
```

```python
plt.show(scatter)
```

## Logarithmic Function

```python
x_array = np.arange(0.01, 10, 0.01)
y_array = np.log(x_array)
```

```python
scatter = tk.make_scatter({
    "x": x_array,
    "y": y_array,
    "mode": "markers",
    "marker": {"maxdisplayed": 50}
  })
plt = tk.plotly()
```

```python
plt.set_title("logarithmic")
plt.set_x_title("horizontal", "x")
plt.set_y_title("vertical", "log(x)")
```

```python
plt.show(scatter)
```

```python
plt.delete_axis_layout("x", "range")
plt.delete_axis_layout("x", "dtick")
```

```python
plt.set_axis_layout("x", "type", "log")
```

```python
plt.show(scatter)
```

## Linear Function

```python
x_array = np.arange(1.0, 5.0, 0.01)
y_array = x_array + 1.0
```

```python
scatter = tk.make_scatter({
    "x": x_array,
    "y": y_array,
    "mode": "markers",
    "marker": {"maxdisplayed": 50}
  })
plt = tk.plotly()
```

```python
plt.set_title("linear")
plt.set_x_title("horizontal", "x")
plt.set_y_title("vertical", "x + 1")
```

```python
plt.set_axis_layout("[xy]", "type", "log")
```

```python
plt.show(scatter)
```
