## Initialization

```python
import tk_plot_utils as tk
```

Note that `tk_plot_utils.pl` is equal to `plotly.offline`, and `tk_plot_utils.go` is equal to `plotly.graph_objs`.

```python
print(tk.pl.__name__)
print(tk.go.__name__)
```

First, we need to initialize Plotly. If there is no Internet connection, we might load local Plotly by setting `connected=False`. However, this loading takes a bit of time and increases notebook size.

```python
tk.init_plotly(connected=True)
```

## Original Plotly way

```python
import numpy as np
xarray = np.arange(-2*np.pi, 2*np.pi, 0.1)
tk.pl.iplot([{"x": xarray, "y": np.sin(xarray)}])
```

## Do the same thing using this module!

```python
#import numpy as np
#xarray = np.arange(-2*np.pi, 2*np.pi, 0.1)
plot = tk.plotly()
trace = tk.make_scatter({"x": xarray, "y": np.sin(xarray)})
plot.show(trace)
```

**Feature**

* clean background
* downloadable as SVG
* fixed size (default is 450x450 in px)
* frame line with major and minor ticks
